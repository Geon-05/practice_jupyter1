```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
import pandas as pd
from konlpy.tag import Okt,Kkma,Hannanum
```

```python
df = pd.read_csv('./전세사기_네이버지식인.csv')
tmp_txt = ' '.join(df['제목'])
```

```python
tmp_txt
```

```python
wc=WordCloud(font_path="C:/windows/fonts/malgun.ttf",
            background_color='black',
            width=500,
            height=500,
            max_words=100,
            max_font_size=200)
            #stopwords = stopwords)
wc=wc.generate(tmp_txt)
wc.to_image()
```

```python
stop_word=['전세사기', '질문','문의','관련','있을까요','전세사기...']
```

```python
# 단어리스트
word_list = tmp_txt.split()
```

```python
word_res = [i for i in word_list if i not in stop_word] # 정제된 단어
doc_res = ' '.join(word_res)
doc_res
```
