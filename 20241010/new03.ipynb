```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import pandas as pd
import os

def search_sigu():
    options=Options()
    options.add_experimental_option('detach',True) #화면이 꺼지지 않고 유지
    options.add_argument('--start-maximized') # 화면을 최대화
    service=Service(ChromeDriverManager().install()) #크롬드라이버 자동 설치
    
    driver=webdriver.Chrome(service=service, options=options)
    
    url='https://www.opinet.co.kr/searRgSelect.do'
    driver.get(url)
    time.sleep(2)
    
    elem_si = driver.find_element(By.XPATH,'//*[@id="SIDO_NM0"]')
    si_list = elem_si.find_elements(By.TAG_NAME,'option')
    elem_gu = driver.find_element(By.XPATH,'//*[@id="SIGUNGU_NM0"]')
    gu_list = elem_gu.find_elements(By.TAG_NAME,'option')
    si_names=[i.text for i in si_list][1:]
    gu_names=[i.text for i in gu_list][1:]
    res = {}
    for i in si_names:
        tmp = []
        for j in gu_names:
            tmp.append(j)
        res[i] = tmp
    driver.quit()
    return res

sigu_dict = search_sigu()

for si in sigu_dict:
    # 다운로드 설정
    download_path = os.path.abspath(f"/Users/go121/practice_jupyter/py/20241010/{si}")  # 다운로드 경로
    # ✨ 저장폴더 만들기
    os.makedirs(download_path, exist_ok=True)
    options = webdriver.ChromeOptions()
    prefs = {
        "download.default_directory": download_path,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "safebrowsing.enabled": True
    }
    options.add_experimental_option("prefs", prefs)
    service=Service(ChromeDriverManager().install())
    
    # 브라우저 시작
    driver=webdriver.Chrome(service=service, options=options)
    
    # 웹사이트 이동 및 파일 다운로드 (예시)
    url='https://www.opinet.co.kr/searRgSelect.do'
    driver.get(url)
    time.sleep(2)

    driver.find_element(By.XPATH,'//*[@id="SIDO_NM0"]').send_keys(si)
    time.sleep(3)
    
    for gu in sigu_dict[si]:
        driver.find_element(By.XPATH,'//*[@id="SIGUNGU_NM0"]').send_keys(gu)
        time.sleep(3)

        print(f"{si}시 {gu} 다운로드를 시작합니다.")
        
        download_button = driver.find_element(By.XPATH,'//*[@id="templ_list0"]/div[7]/div/a')
        download_button.click()
        
        # 파일 다운로드를 기다림
        time.sleep(10)  # 파일 다운로드 대기 (다운로드 시간에 맞게 설정)
        
        # 파일명 변경
        original_filename = os.path.join(download_path, "지역_위치별(주유소).xls")
        new_filename = os.path.join(download_path, f"{si}시_{gu}_주유소.xls")
        
        # 파일이 존재하는지 확인 후 이름 변경
        if os.path.exists(original_filename):
            os.rename(original_filename, new_filename)
    print('모든 파일이 다운로드 되었습니다.')
    driver.quit()
```
