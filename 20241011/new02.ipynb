```python
txtList = [i.split() for i in txt]
txtList
```

```python
# Î¨∏ÏÑú Ï†ÑÏ≤¥Ïóê Îì±Ïû•ÌïòÎäî Îã®Ïñ¥Îì§
nouns = sorted(list(set([j for i in txtList for j in i])))
nouns
```

```python
dtm = []
for i in txtList:
    dtm.append([])
    for j in nouns:
        dtm[-1].append(i.count(j))
dtm
# ü§îÏïÑÎûò Ï≤òÎüº ÌëúÌòÑÌï¥ÎèÑÎê®!
# dtm = []
# for i in txtList:
#     tmp = []
#     for j in nouns:
#         tmp.append(i.count(j))
#     dtm.append(tmp)
# dtm
```
