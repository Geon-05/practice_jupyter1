```python
import numpy as np
import pandas as pd
txt=['파이썬 차트 파이썬 머신러닝',
      '차트 파이썬 R 차트',
       'R 분석 시각화'
      ]
```

```python
txtList = [i.split() for i in txt]
txtList
```

```python
# 문서 전체에 등장하는 단어들
nouns = sorted(list(set([j for i in txtList for j in i])))
nouns
```

```python
dtm = []
for i in txtList:
    dtm.append([])
    for j in nouns:
        dtm[-1].append(i.count(j))
dtm
# 🤔아래 처럼 표현해도됨!
# dtm = []
# for i in txtList:
#     tmp = []
#     for j in nouns:
#         tmp.append(i.count(j))
#     dtm.append(tmp)
# dtm
```

```python
tf = pd.DataFrame(dtm,columns=nouns)
tf
```

```python
# bool타입으로 바꾼다!
a = [3,4,0,5,1]
[bool(i) for i in a]
```

```python
df = tf.astype('bool').sum()
df
```

```python
D = len(tf)
D
```

```python
idf = D/df
idf
```

```python
# TF-IDF
tfidf_ = tf*idf
tfidf_
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(txt)
tfidf_vectorizer
```

```python
tfidf_feature = tfidf_vectorizer.get_feature_names_out()
tfidf_feature
```
