```python
N = 100000
x = np.random.randn(N) # 0, 1 ()
plt.subplot(1,2,1)
plt.hist(x, bins=10)
plt.subplot(1,2,2)
plt.hist(x, bins=100)
plt.show()
```

```python
N = 100000
x = np.random.randn(N) # 0, 1 ()
plt.subplot(1,2,1)
plt.hist(x, bins=10)
plt.ylim(0,30000)
plt.subplot(1,2,2)
plt.hist(x, bins=100)
plt.ylim(0,30000)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.hist(x,bins=100,density=True)
plt.subplot(1,2,2)
plt.hist(x,bins=100,density=True, cumulative=True)
plt.show
```
