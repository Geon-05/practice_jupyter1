```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df = pd.read_csv('https://bit.ly/ds-house-price-clean')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df.index, df['분양가'], alpha=0.3)
```

```python
jis = df['지역'].unique()
for i in jis:
    tmp = df[df['지역']==i].reset_index(drop=True)
    # display(tmp)
    plt.scatter(tmp.index, tmp['분양가'])
    plt.show()
```

```python
jis = df['지역'].unique()
gus = df['규모'].unique()
for i in jis:
    for j in gus:
        con1 = df['지역']==i
        con2 = df['규모']==j
        tmp = df[con1 & con2].reset_index(drop=True)
        plt.figure(figsize=(10,3))
        plt.subplot(1,2,1)
        plt.scatter(tmp.index,tmp['분양가'])
        plt.title(f'{i}_{j}_산점도')
        plt.subplot(1,2,2)
        tmp['분양가'].plot(kind='box')
        plt.title(f'{i}_{j}_박스플롯')
        plt.tight_layout()
        plt.show
```
