```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from collections import Counter
```

```python
url = 'https://news.naver.com/section/105'

html = requests.get(url)

soup = bs(html.text, 'html')
```

```python
news_title = [i.text for i in soup.select('strong.sa_text_strong')]
news_title
```

```python
res = []
for i in news_title:
    clear = i.replace('\\','').replace('[','').replace(']','').replace('\'','').replace('\"','').replace('“','').replace('”','').replace('`','')
    clear = clear.replace(',','').replace('...','').replace('…','').replace('‘','').replace('’','').replace('?','').replace('!','').replace('(',' ').replace(')','')
    word_list = clear.split(" ")
    for j in word_list:
        print(j)
        res.append(j)

Counter(res).most_common(10)
```
