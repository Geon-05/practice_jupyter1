```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
import pandas as pd
import json

client_id = "sINqNPstJrYv4dDyStMK"
client_secret = "SKpquTbQWh"

def naver_search(search, n_display=100, start=1, server_name='blog'):
    res = []
    while n_display>0:
        if n_display > 100:
            base_url = f'https://openapi.naver.com/v1/search/{server_name}'
            p = {
                'query':search,
                'display':100,
                'start':start
            }
        else :
            base_url = f'https://openapi.naver.com/v1/search/{server_name}'
            p = {
                'query':search,
                'display':n_display,
                'start':start
            }
        enc_params = urllib.parse.urlencode(p)
        url = f'{base_url}?{enc_params}'
        
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if(rescode==200):
            response_body = response.read()
            res_json = json.loads(response_body)
            df = pd.DataFrame(res_json['items'])
            res.append(df)
        else:
            print("Error Code:" + rescode)
            
        n_display -= 100
        start +=100
        
    res_df = pd.concat(res, ignore_index=True)
    return res_df
```

```python
app_dict = {
    '블로그':'blog',
    '뉴스':'news',
    '카페':'cafearticle',
    '쇼핑':'shop',
    '지식인':'kin'
}

for i in app_dict.values():
    res = naver_search(search="스타벅스", n_display=200, server_name=i)
    res['server'] = i
    print(f'다음은 {i}의 검색결과입니다.')
    display(res)
```
