```python
import folium
import pandas as pd
```

```python
popup_text = '<div style="width: 50px;">대전시청</div>'
# 지도차트를 그렸음
map = folium.Map(location=[36.3504567, 127.3848187], zoom_start=7)
folium.Marker(
  location=[36.3504567, 127.3848187],
  popup=popup_text,
  icon=folium.Icon(icon='glyphicon glyphicon-music')
).add_to(map)
folium.Marker(
  location=[37.5665851, 126.9782038],
  popup='<div style="width: 50px;">서울시청</div>',
  icon=folium.Icon(color='red', icon='glyphicon glyphicon-headphones')
).add_to(map)
folium.Marker(
  location=[35.179665, 129.0747635],
  popup='<div style="width: 50px;">부산시청</div>',
  icon=folium.Icon(color='orange', icon='glyphicon glyphicon-tag')
).add_to(map)
map

# 서울시청 37.5665851!4d126.9782038, 부산시청35.179665!4d129.0747635
```

```python
df = pd.read_excel('./crim_1.xlsx')
```

```python
df
```

```python
df.dtypes
```

```python
lank3_index = df.sort_values('절도',ascending=False).head(3).index
lank3_index
```

```python
for i in lank3_index:
    print(i)
```

```python
map = folium.Map(location=[37.5665851, 126.9782038], zoom_start=11)

for idx, i in enumerate(lank3_index):
    tmp = df['자치구'][i]
    tmp_n = df['절도'][i]
    popup_text = f'<div style="width: 70px;">[{idx+1}등]<br>{tmp}<br>{tmp_n}명</div>'
    folium.Marker([df['위도'][i], df['경도'][i]],
                 popup=popup_text,
                 icon=folium.Icon(icon='glyphicon glyphicon-music')
                 ).add_to(map)

    folium.CircleMarker([df['위도'][i], df['경도'][i]],
                         radius=int(tmp_n)/30,
                         color='#fffggg',
                         fill_color='#fffggg').add_to(map)
map
```

```python
# 서울 절도가 가장 많이 잡은 경찰서 3군데 원그래프로 지도차트 그래세요.
sort_df = df.sort_values('절도', ascending=False, ignore_index=True)
```

```python
sort_rank_df = sort_df.iloc[:3,:]
```

```python
map = folium.Map(location=[37.5665851, 126.9782038], zoom_start=11)
for i in sort_rank_df.values:
    loca = [i[8], i[9]]
    name = i[1]
    crim = i[5]
    folium.Marker(
      location=loca,
      popup=name,
      icon=folium.Icon(icon='glyphicon glyphicon-music')
    ).add_to(map)
    folium.CircleMarker(loca,
                       radius=crim/50,
                       color='#fffggg',
                       fill_color='#fffggg').add_to(map)
map
```
