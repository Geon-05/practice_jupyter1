```python
import pandas as pd
import requests
```

```python
def serch_code(areaCode):
    url='http://apis.data.go.kr/B551011/KorService1/areaCode1'
    param={'serviceKey':'kzCpdqIG1LWIVrQymfmp3+JWFtCzoOkxDatyr3THUvTMPt0TZE2Yldkc+JOnhG60HUUjkjz89Y2LpIj4pvpP2w==',
        'numOfRows':'1000',
        'pageNo': '1',
        'MobileOS':'ETC',
        'MobileApp':'AppTest',
        'areaCode' : areaCode,
        '_type':'json'
          }
    response = requests.get(url, params=param).json()
    res = pd.DataFrame(response['response']['body']['items']['item'])[['code','name']]
    return res
```

```python
si_df = serch_code('')
si_df
```

```python
res = []
for si_code, si_name in si_df.values:
    tmp = serch_code(si_code)
    tmp[['si_code', 'si_name']] = si_code, si_name
    res.append(tmp)
code_df = pd.concat(res, ignore_index=True)
```
