```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df = sns.load_dataset('tips')
print(df.shape)
df.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='size')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='day')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, row='time')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, row='day')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, col='day')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df, row='day', col='sex')
plt.show()
```
