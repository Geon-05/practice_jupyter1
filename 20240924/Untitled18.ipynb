```python
from PIL import Image
from PIL import ImageFilter
from glob import glob
import os
import re
import shutil

file_names = glob('./images/*.jpg')


filter_list = [ImageFilter.BLUR, ImageFilter.CONTOUR, ImageFilter.DETAIL,
               ImageFilter.EDGE_ENHANCE, ImageFilter.EDGE_ENHANCE_MORE, ImageFilter.EMBOSS, ImageFilter.FIND_EDGES,
               ImageFilter.SHARPEN, ImageFilter.SMOOTH, ImageFilter.SMOOTH_MORE]

for file_name in file_names:
    img = Image.open(file_name)
    just_file_name =re.findall("[가-힣]+",file_name)[0] #강아지, 고양이 .. 등등
    for i in filter_list:
        filter_img = img.filter(i) #필터씌운 이미지
        filter_name = str(i).split('.')[-1].strip("'>'") #필터 이름 변수로 저장
        os.makedirs(filter_name,exist_ok=True) #필터 이름으로 폴더 만들기
        new_path = file_name.replace("./images",f"{filter_name}\\{just_file_name}")
        new_path_file = f"{filter_name}\\{just_file_name}" #필터\\강아지 , 필터\\고양이
        os.makedirs(new_path_file,exist_ok=True)
        shutil.copyfile(file_name,new_path)
```
