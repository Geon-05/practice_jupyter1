```python
from glob import glob
from PIL import Image
import pandas as pd
```

```python
# 1. 파일 목록 가져오기 (glob)
file_names = glob('./images/*.jpg')
# print(file_names)
res = []

# 2. 이미지 읽어서 속성을 가져오기 (pillow)
for file_name in file_names:
    tmp = {}
    img = Image.open(file_name)
    tmp['파일명'] = img.filename
    tmp['크기'] = img.size
    tmp['가로크기'] = img.width
    tmp['세로크기'] = img.height
    tmp['모드'] = img.mode
    tmp['포멧'] = img.format
    res.append(tmp)
    # print(tmp)
    # print('-'*30)
# print(res)
# 3. 데이터 프레임으로 바꾸어 csv로 저장하기 (pandas)
df = pd.DataFrame(res)
# display(df) # 주피터노트북에서만 쓰이는 명령어 display
df.to_csv('이미지정보.csv')
```
