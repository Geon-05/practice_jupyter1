```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import random
'''
미션
korea('전세사기',12) => 12페이지까지
korea('전세사기',0) => 마지막 페이지까지
'''
options=Options()
options.add_experimental_option('detach',True)
options.add_argument('--start-maximized')
service=Service(ChromeDriverManager().install())

driver=webdriver.Chrome(service=service, options=options)

def korea(search, pages=1):
    # ✨ 기본 검색 url 설정
    url = 'https://www.korea.kr/totalSearch.do?pWise=main&pWiseMain=S1&where=ALL&multiFlag=Y&dateRange=all&range=ALL&multiFlag=Y&topSearchKeyword=+'
    
    # ✨ 접속 및 검색
    driver.get(url)
    time.sleep(1)
    elem = driver.find_element(By.XPATH,'//*[@id="topSearchKeywordSr"]')
    elem.send_keys(search)
    elem.send_keys(Keys.ENTER)
    time.sleep(1)
    driver.find_element(By.XPATH,'//*[@id="news_korea_2023"]/div[1]/div/ul/li[2]/a').click()
    time.sleep(1)
    html = driver.page_source
    res = [html]
    
    # ✨ 마지막 페이지 번호 조회
    max_page_elem = driver.find_element(By.XPATH,'//*[@id="t_contanier"]/div/div[2]/div[1]/div[1]/strong[2]')
    max_page = int(int(max_page_elem.text.replace('건','').replace(',',''))/10+1)
    
    # ✨ 사용자 요청 페이지와 마지막 페이지 번호 비교
    if (pages > max_page) or (pages == 0):
        pages = max_page

    # ✨ 현재 url 추출 
    now_url = driver.current_url

    # ✨ 2페이지 이후 검색 
    for i in range(2,pages+1):
        new_url = now_url.replace('scnt=1',f'scnt={i}')
        driver.get(new_url)
        # 🤔 2~4까지의 랜덤한 실수 반환하여 봇 아닌척하기!
        time.sleep(random.uniform(2,4))
        res.append(driver.page_source)
    
    # ✨ 드라이버 종료 및 html 리스트 반환
    driver.quit()
    return res

# print(len(korea('부동산',0)))
# 👍-> 0의경우 193페이지를 크롤링하여 주석처리함
print(korea('부동산',2))
```
