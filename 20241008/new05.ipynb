```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
search = '파이썬'
page = 1
url = f'https://kin.naver.com/search/list.naver?query={search}&page={page}'
html = requests.get(url)
soup = bs(html.text, 'html')

타이틀 = [i.text for i in soup.find_all('a','_nclicks:kin.txt _searchListTitleAnchor')]
링크 = [i['href'] for i in soup.find_all('a','_nclicks:kin.txt _searchListTitleAnchor')]
날짜 = [i.text for i in soup.find_all('dd', 'txt_inline')]
요약글 = [i.text for i in soup.select('ul.basic1 dl dd')[1::4]]

res = []
for i in range(len(타이틀)):
    tmp = {}
    tmp['타이틀'] = 타이틀[i]
    tmp['날짜'] = 날짜[i]
    tmp['요약글'] = 요약글[i]
    tmp['링크'] = 링크[i]
    res.append(tmp)
df = pd.DataFrame(res)
df
```

```python
def search_kin(search, page=1):
    res = []
    for real_page in range(1,page+1):
        url = f'https://kin.naver.com/search/list.naver?query={search}&page={page}'
        html = requests.get(url)
        soup = bs(html.text, 'html')
        
        타이틀 = [i.text for i in soup.find_all('a','_nclicks:kin.txt _searchListTitleAnchor')]
        링크 = [i['href'] for i in soup.find_all('a','_nclicks:kin.txt _searchListTitleAnchor')]
        날짜 = [i.text for i in soup.find_all('dd', 'txt_inline')]
        요약글 = [i.text for i in soup.select('ul.basic1 dl dd')[1::4]]
        
        for i in range(len(타이틀)):
            tmp = {}
            tmp['타이틀'] = 타이틀[i]
            tmp['날짜'] = 날짜[i]
            tmp['요약글'] = 요약글[i]
            tmp['링크'] = 링크[i]
            res.append(tmp)
            
    df = pd.DataFrame(res)
    return df
```
