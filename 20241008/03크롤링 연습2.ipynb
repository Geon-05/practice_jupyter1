```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://kin.naver.com/search/list.naver?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&page=1'
html = requests.get(url)
html
```

```python
soup = bs(html.text, 'html')
soup
```

```python
titles = soup.find_all('a','_nclicks:kin.txt _searchListTitleAnchor')
titles
```

```python
res = []
for i in titles:
    tmp = {}
    tmp['제목'] = i.text
    tmp['링크'] = i['href']
    res.append(tmp)
pd.DataFrame(res)
```
