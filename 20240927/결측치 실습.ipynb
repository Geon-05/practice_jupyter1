```python
'''
판다스객체.fillna()
1. 특정값으로 채우기 : 0, 평균, 중앙값, 최빈값...
2. 위쪽 데이터, 아래쪽 데이터, 특정 컬럼의 데이터
'''
```

```python
df.fillna(0)
```

```python
df['키'].mean()
```

```python
df['키1'] = df['키'].fillna(df['키'].mean())
df
```

```python
df['키2'] = df['키'].fillna(df.groupby('성별')['키'].mean()['여자'])
df
```

```python
df['그룹1'] = df['그룹'].fillna(df['그룹'].value_counts().head(1).index[0])
df
```

```python
df['그룹'].fillna(method='ffill')
```

```python
df['그룹'].fillna(method='bfill')
```

```python
df['그룹2'] = df['그룹'].fillna(df['이름'])
df
```
