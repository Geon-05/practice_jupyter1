```python
df = pd.read_csv('./data/seoul_house_price.csv')
print(df.shape)
df.head()
```

```python
df.columns = ['지역', '규모', '연도', '월', '분양가격']
```

```python
df
```

```python
# 분양가격에 에러나는 모든 데이터를 수집하세요 (try~except~)
res = []
for i in list(df['분양가격']):
    try :
        res.append(int(i))
    except :
        res.append(None)
print(res)
df['분양가격'] = res
```

```python
res = []
for i in df['분양가격']:
    # print(i, type(i))
    try :
        tmp = int(i)
    except :
        res.append(i)

list(set(res))
```

```python
res = []
for i in df['분양가격']:
    try:
        tmp = int(i.replace(',',''))
        res.append(tmp)
    except:
        res.append(None)
df['분양가격'] = res
```

```python
df = df[df['규모']!='전체'].reset_index(drop=True)
df
```

```python
res = []
for i in df['규모']:
    i = i.replace('㎡','')
    res.append(i)
df['규모'] = res
```

```python
res = []
for i in df['월']:
    if i in [1,2,3]:
        res.append("1분기")
    elif i in [4,5,6]:
        res.append("2분기")
    elif i in [7,8,9]:
        res.append("3분기")
    elif i in [10,11,12]:
        res.append("4분기")
df['분기'] = res
df
```

```python
res = []
for i in df['지역']:
    if i in ['서울', '경기', '인천']:
        res.append("수도권")
    else :
        res.append("지방")
df['권역'] = res
df
```

```python
df.isnull().sum()
```

```python
season = {
    '1분기':'봄',
    '2분기':'여름',
    '3분기':'가을',
    '4분기':'겨울'
}
season
```

```python
df['계절'] = df['분기'].map(season)
df
```
