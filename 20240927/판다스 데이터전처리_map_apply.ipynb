```python
df.columns = ['지역', '규모', '연도', '월', '분양가격']
```

```python
df.head()
```

```python
# 분양가격 컬럼을 int로 바꾸기 -> ,는 삭제하고, 나머지는 None으로 하기
def number_t(value:str):
    try :
        res = int(value.replace(',',''))
    except:
        res = None
    return res

# 1. for문 -1
res = []
for i in df['분양가격']:
    res.append(number_t(i))
print(res)
```

```python
# 2. for문 -2
res1 = [number_t(i) for i in df['분양가격']]
print(res1)
```

```python
# 3. map
df['분양가격'].map(number_t)
```

```python
df.head()
```

```python
df['규모'].map(lambda x : x.replace('㎡',''))
```

```python
df[['연도','월']] = df[['연도','월']].map(lambda x : f'{x}원')
df[['연도','월']] = df[['연도','월']].map(lambda x : int(x.replace('원','')))
```

```python
df.dtypes
```

```python
df.head()
```

```python
res = []
for i in df[['연도','월']]:
    tmp = []
    for j in df[i]:
        tmp.append(str(j)+'원')
    res.append(tmp)
df[['연도','월']] = pd.DataFrame(res).T
```
