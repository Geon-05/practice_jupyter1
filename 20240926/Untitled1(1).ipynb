```python
'''
# 1. '부동산'폴더 만들기
# 2. '부동산'폴더에 지역별 csv로 저장하기 (./부동산/서울.csv, ./부동산/경기.csv)
    - df의 '지역'컬럼 유니크뽑아라
'''

import os
import pandas as pd

df = pd.read_csv('./data/house_price_clean.csv')
os.makedirs('./부동산',exist_ok=True)

for site in df['지역'].unique():
    temp = df[df['지역']==site].reset_index(drop=True)
    temp.to_csv(f'./부동산/{site}.csv', encoding='cp949')
```
