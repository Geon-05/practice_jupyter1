```python
import pandas as pd
df = pd.read_csv('./data/house_price_clean.csv')
df
```

```python
# max_ol => max_ol_df
def max_outlier(df, col):
    mn, mx = df[f'{col}'].quantile([0.25,0.75])
    IQR = mx - mn
    max_ol = mx + (IQR * 1.5)
    max_ol_df = df[df[f'{col}'] > max_ol].reset_index(drop=True)
    return max_ol_df

# min_ol => min_ol_df
def min_outlier(df, col):
    mn, mx = df[f'{col}'].quantile([0.25,0.75])
    IQR = mx - mn
    min_ol = mn - (IQR * 1.5)
    min_ol_df = df[df[f'{col}'] < min_ol].reset_index(drop=True)
    return min_ol_df

# normal => normal_df
def normal_outlier(df, col):
    mn, mx = df[f'{col}'].quantile([0.25,0.75])
    IQR = mx - mn
    max_ol = mx + (IQR * 1.5)
    min_ol = mn - (IQR * 1.5)
    con1 = df[f'{col}'] <= max_ol
    con2 = df[f'{col}'] >= min_ol
    normal_df = df[con1 & con2].reset_index(drop=True)
    return normal_df

display(max_outlier(df,'분양가'))
display(min_outlier(df,'분양가'))
display(normal_outlier(df,'분양가'))

# class Outlier:
#     def __init__(self, df, col):
#         self.df = df
#         self.col = col
```
