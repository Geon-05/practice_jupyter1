```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
df = pd.read_csv('./data/cctv_in_seoul.csv')
print(df.shape)
df.head()
```

```python
df['지역']='서울'
```

```python
df
```

```python
df['2016년 증감']=df['2016년']-df['2015년']
```

```python
df
```

```python
res = []
for i in df['2016년 증감']:
    if i >=0 :
        res.append('증가')
    else :
        res.append('감소')
df['기타'] = res
df.head()
```

```python
res = ['증가' if i >= 0 else '감소' for i in df['2016년 증감']]
df['비고1'] = res
df.head()
```

```python
df['2016년'].describe()
```

```python
res = []
for i in df['2016년']:
    if i <= 292:
        tmp = '하'
    elif i <= 467:
        tmp = '중'
    else :
        tmp = '상'
    res.append(tmp)
df['구분']=res
df
```
